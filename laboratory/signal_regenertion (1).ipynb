```python
#Unit sample sequence code



import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Step 2: Define parameters
n = 11  # Number of samples
n0 = 5  # Offset (Shift index)

# Step 3: Generate unit impulse sequence
imp = signal.unit_impulse(n, n0)  # Generate unit impulse

# Step 4: Plotting the unit sample sequence
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.stem(np.arange(0, n), imp, 'r', markerfmt='ro', basefmt='k')
plt.xlabel('Sample #')
plt.ylabel(r'$\delta[n-n_0]$')
plt.title('Discrete-time Unit Impulse')
plt.show()
```

```python

#Unit ramp signal code

import numpy as np
import matplotlib.pyplot as plt

# Define range of n from -2 to 5
n = np.arange(-2, 6, 1)
y = []

# Compute unit ramp signal values
for i in range(len(n)):
    temp = (n[i] if n[i] >= 0 else 0)
    y.append(temp)

# Print values for verification
print("n:", list(n))
print("Unit Ramp Signal y:", y)

# Plotting the graph
plt.stem(n, y)
plt.axis([-2.1, 5.1, -0.1, max(y)+1])
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Unit Ramp Signal')
plt.grid(True)
plt.show()
```

```python
#exponential code

import numpy as np
import matplotlib.pyplot as plt

a = 0.9  # Decay factor
n = range(0, 26)  # Range of n values from 0 to 25

# Generate the exponential signal
y = [a ** i for i in n]

# Plotting the graph
plt.stem(n, y)
plt.axis([-0.5, 25.5, -0.1, 1.1])
plt.xlabel('n --->')
plt.ylabel('Amplitude ---->')
plt.title('Exponential Signal for 0 < a < 1')
plt.grid(True)
plt.show()
```

```python
#	Triangular signal

def triangle(length, amplitude):
    section = length // 4
    for direction in (1, -1):
        for i in range(section):
            yield i * (amplitude / section) * direction
        for i in range(section):
            yield (amplitude - (i * (amplitude / section))) * direction

# Generate an array of length 100 with max/min amplitude of 0.5
triangle_wave_custom = list(triangle(100, 0.5))

# Plotting the custom triangle wave
plt.plot(triangle_wave_custom)
plt.title('Custom Triangle Wave')
plt.xlabel('Sample Number')
plt.ylabel('Amplitude')
plt.grid()
plt.show()





```
